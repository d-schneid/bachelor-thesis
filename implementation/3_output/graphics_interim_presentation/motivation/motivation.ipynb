```python
import matplotlib as mpl
from matplotlib import font_manager as fm, rcParams
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
```

```python
fpath = os.path.join(mpl.get_data_path(), r"C:\Users\danie\Documents\Karlsruhe\03_KIT_Studium\07_Semester\Bachelorarbeit\Bachelorarbeit\Repo\implementation\3_output\fonts\lato\Lato-Regular.ttf")
prop = fm.FontProperties(fname=fpath, size=20)
blue_color = "#0E2356"
green_color = "#008052"
labelsize = 12

# Create some data
dates = pd.date_range(start='2022-01-01 00:00:00', end='2022-01-02 00:00:00', freq='H')
np.random.seed(1)
values = np.random.uniform(low=15, high=30, size=len(dates))

# Combine the dates and values into a DataFrame
df = pd.DataFrame({'dates': dates, 'values': values})

# Set the date column as the index
df.set_index('dates', inplace=True)

# Plot the time series data
plt.figure(figsize=(15, 5))
plt.plot(df.index, df['values'], color=green_color)

# Specify axes ticks
plt.xticks(pd.date_range(start='2022-01-01 00:00:00', end='2022-01-02 00:00:00', freq='H'), [d.strftime("%H:%M") for d in dates], rotation=45, fontproperties=prop, color=blue_color)
plt.yticks(fontproperties=prop, color=blue_color)

# Add labels and title to the plot
labelpad = 12
pad = labelpad
plt.xlabel('Uhrzeit', fontproperties=prop, color=blue_color, labelpad=labelpad)
plt.ylabel('Temperatrur in C', fontproperties=prop, color=blue_color, labelpad=labelpad)
plt.title('Werte eines Temperatursensors innerhalb eines Tages', fontproperties=prop, color=blue_color, pad=pad)

# Get the axis object
ax = plt.gca()

# Set the color of the spines (the box surrounding the plot)
for spine in ax.spines.values():
    spine.set_edgecolor(blue_color)

# Set the color of the tick marks and labels
ax.tick_params(axis='both', color=blue_color, labelsize=labelsize)

plt.savefig("temperature.svg", bbox_inches='tight')

# Show the plot
plt.show()
```

```python
def normalize(time_series):
    normalized_time_series = [100.0]
    first_value = time_series[0]
    for i in range(1, len(time_series)):
        normalized_time_series.append(time_series[i] / first_value * 100)
    return normalized_time_series
```
